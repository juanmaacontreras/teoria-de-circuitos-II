```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Definimos la transferencia normalizada T(sn) = (sn - 1) / (sn + 1)
# Coeficientes: numerador [1, -1], denominador [1, 1]
num = [1, -1]
den = [1, 1]

sys = signal.TransferFunction(num, den)

# Generamos el diagrama de Bode
w = np.logspace(-2, 2, 1000) # De 0.01 a 100 rad/s
w, mag, phase = signal.bode(sys, w)

# Graficamos
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# Módulo
ax1.semilogx(w, mag)
ax1.set_title('Diagrama de Bode - Filtro Pasa-Todo Normalizado')
ax1.set_ylabel('Amplitud [dB]')
ax1.set_ylim(-1,1)
ax1.grid(True, which="both")

# Fase
ax2.semilogx(w, phase)
ax2.set_ylabel('Fase [grados]')
ax2.set_xlabel('Frecuencia normalizada [rad/s]')
ax2.grid(True, which="both")

plt.tight_layout()
plt.show()
```
